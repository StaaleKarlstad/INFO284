```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('Datasets/elektronisk-rapportering-ers-2018-fangstmelding-dca-simple.csv', delimiter=';')
```

```python
# Starttidspunkt og Stopptidspunkt inneholder både dato og klokkeslett. Dette gjør en del andre kolonner redunante.
# Vi kan også finne varighet ved Stopptidspunkt - Starttidspunkt
# Fangstår virker å være lik årstallet i Stopptidspunkt
# Kanskje vi burde beholde kolonner med koder og opprette dictionaries hvor vi kan slå opp tilhørende art/redskap

new_df = df.drop(columns = ['Meldingsdato', 'Meldingsklokkeslett', 'Startdato', 'Startklokkeslett', 'Hovedområde start (kode)', 'Stoppdato',
                            'Stoppklokkeslett', 'Varighet', 'Fangstår', 'Redskap FAO (kode)', 'Redskap FDIR (kode)', 'Hovedart FAO (kode)', 'Art FAO (kode)', 
                            'Art - FDIR (kode)', 'Art - gruppe (kode)', 'Lengdegruppe (kode)'] )
```

```python
# Dropper en hel rad hvor en gitt kolonne inneholder en spesifikk verdi

#     df = df.drop(df[df['column_name'] == 'specific_value'].index)

# Hvis viktige felter ikke er fyllt ut, kan vi da stole på at resten av fangstmeldingen gir oss korrekt informasjon? 

# Dropper en hel rad hvor en gitt kolonne inneholder en nullverdi
new_df = new_df.dropna(subset = ['Hovedområde start'])

# dropna() kan brukes uten innsente argumenter for å fjerne rader med nullverdier
# Feltene for 'Bruttotonnasje 1969' og 'Bruttotonnasje annen' vil som regel ikke være fyllt ut samtidig.
# Derfor vil de fleste radene i tabellen inneholde minst én null-verdi
```

```python
new_df['Starttidspunkt'] = pd.to_datetime(df['Starttidspunkt'])
new_df['Stopptidspunkt'] = pd.to_datetime(df['Stopptidspunkt'])
new_df['Meldingstidspunkt'] = pd.to_datetime(df['Meldingstidspunkt'])
```
